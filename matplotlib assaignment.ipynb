 Q4) What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.



Bar Graphs are the pictorial representation of the data present. It is a way to analyze the data in a pictorial format. 

```python
import matplotlib.pyplot as plt
letters= ['A', 'B', 'C', 'D', 'E']
values = [3, 7, 1, 8, 4]
plt.bar(letters, values)
plt.title('Bar Plot')
plt.show()
```

Q5)What is a box plot? Why is it used? Using the following data plot a box plot.


Box Plot: It is a type of chart that depicts a group of numerical data through their quartiles. It is a simple way to visualize the shape of our data.



 


```python
data = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]
plt.boxplot(data)
plt.title('Box Plot Example')
plt.ylabel('Values')
plt.show()
```
